```python
# -------- Functions --------
def upload_image():
    global img

    file_path = filedialog.askopenfilename(
        filetypes=[("Image Files", "*.jpg *.png *.jpeg")]
    )

    if file_path:
        img = cv2.imread(file_path)
        show_image(img,panel_original)


def convert_to_sketch():
    global img, sketch

    if img is None:
        return

    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    invert = 255 - gray
    blur = cv2.GaussianBlur(invert, (21, 21), 0)
    inverted_blur = 255 - blur
    sketch = cv2.divide(gray, inverted_blur, scale=256.0)

    show_image(sketch, panel_sketch, is_gray=True)


def save_image():
    global sketch

    if sketch is None:
        return

    save_path = filedialog.asksaveasfilename(defaultextension=".jpg")
    if save_path:
        cv2.imwrite(save_path, sketch)


def show_image(image, panel, is_gray=False):
    image = cv2.resize(image, (350, 350))

    if not is_gray:
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    img_pil = Image.fromarray(image)
    img_tk = ImageTk.PhotoImage(img_pil)

    panel.config(image=img_tk)
    panel.image = img_tk

```

```python
import cv2
from tkinter import Tk, filedialog

# Hide main tkinter window
Tk().withdraw()

# Open file dialog
file_path = filedialog.askopenfilename(
    title="Select an Image",
    filetypes=[("Image Files", "*.jpg *.jpeg *.png")]
)

# Check if user selected file
if not file_path:
    print("No file selected!")
    exit()

# Read image
img = cv2.imread(file_path)

# Resize for better display
img = cv2.resize(img, (500, 500))

# ---- Sketch Conversion ----
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
invert = 255 - gray
blur = cv2.GaussianBlur(invert, (21, 21), 0)
inverted_blur = 255 - blur
sketch = cv2.divide(gray, inverted_blur, scale=256.0)
# Convert sketch to 3-channel
sketch = cv2.cvtColor(sketch, cv2.COLOR_GRAY2BGR)

# Combine images
combined = cv2.hconcat([img, sketch])

# Show result
cv2.imshow("Original vs Sketch", combined)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
