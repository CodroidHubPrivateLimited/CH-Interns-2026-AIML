```python
#hand detect and tracking 

mpHands = mp.solutions.hands
hands = mpHands.Hands(
    max_num_hands=1,
    min_detection_confidence=0.7,
    min_tracking_confidence=0.7
)
```

<h1 style="color:#00ffcc; font-family:'times new roman', Tahoma, Geneva, Verdana, sans-serif;">
        Hand Detection & Tracking
    </h1>
    <p style="color:#ffffff; font-size:18px;">
        Initializes the hand tracking model to detect and follow hand movements in real-time.
    </p>

   <hr style="border: 0.5px solid #444;">

   <div style="text-align:left; color:#e0e0e0;">
        <ul>
            <li><b>mp.solutions.hands:</b> Loads MediaPipe hand tracking module</li>
            <li><b>max_num_hands=1:</b> Detects only one hand for better performance</li>
            <li><b>min_detection_confidence=0.7:</b> Ensures accurate hand detection</li>
            <li><b>min_tracking_confidence=0.7:</b> Maintains stable tracking across frames</li>
        </ul>

```python
mpDraw = mp.solutions.drawing_utils
cap = cv2.VideoCapture(0)

last_action = time.time()
```

<div style="padding:35px; background-color:#1a1a1a; border-radius:15px; border: 3px solid #00ffcc; text-align:center;">

 <h1 style="color:#00ffcc; font-family:'times new roman', Tahoma, Geneva, Verdana, sans-serif;">
        Drawing & Camera Setup
    </h1>

 <p style="color:#ffffff; font-size:18px;">
        Sets up hand landmark drawing, webcam input, and action timing control.
    </p>

 <hr style="border: 0.5px solid #444;">

 <div style="text-align:left; color:#e0e0e0;">
        <ul>
            <li><b>drawing_utils:</b> Used to draw hand landmarks and connections</li>
            <li><b>VideoCapture(0):</b> Starts webcam for real-time video input</li>
            <li><b>last_action:</b> Stores last gesture time to prevent repeated triggers</li>
        </ul>
    </div>

</div>

```python
def fingers_up(lmList):
    tips = [4, 8, 12, 16, 20]
    fingers = []

    if lmList[tips[0]][0] > lmList[tips[0]-1][0]:
        fingers.append(1)
    else:
        fingers.append(0)

    for i in range(1, 5):
        if lmList[tips[i]][1] < lmList[tips[i]-2][1]:
            fingers.append(1)
        else:
            fingers.append(0)

    return fingers

```

<div style="padding:35px; background-color:#1a1a1a; border-radius:15px; border: 3px solid #00ffcc; text-align:center;">

 <h1 style="color:#00ffcc; font-family:'times new roman', Tahoma, Geneva, Verdana, sans-serif;">
        Finger Detection Logic
    </h1>

 <p style="color:#ffffff; font-size:18px;">
        Determines which fingers are <b>UP</b> or <b>DOWN</b> using hand landmark positions.
    </p>

 <hr style="border: 0.5px solid #444;">

<div style="text-align:left; color:#e0e0e0;">
        <ul>
            <li><b>tips:</b> Landmark indices of all fingertips</li>
            <li><b>Thumb:</b> Checked using horizontal (x-axis) position</li>
            <li><b>Other Fingers:</b> Checked using vertical (y-axis) position</li>
            <li><b>Logic:</b> Tip above joint → Finger is UP (1), else DOWN (0)</li>
            <li><b>Output:</b> Returns a list like [0,1,0,0,0]</li>
        </ul>
    </div>

</div>

```python
while True:
    success, frame = cap.read()
    if not success:
        continue

    # BLACK SCREEN
    img = np.zeros_like(frame)

    imgRGB = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    results = hands.process(imgRGB)

    if results.multi_hand_landmarks:
        for handLms in results.multi_hand_landmarks:
            lmList = []

            for id, lm in enumerate(handLms.landmark):
                h, w, c = frame.shape
                cx, cy = int(lm.x * w), int(lm.y * h)
                lmList.append((cx, cy))

            fingers = fingers_up(lmList)
            current_time = time.time()

            # NEXT
            if fingers == [0,1,0,0,0] and current_time - last_action > 1:
                pyautogui.press("right")
                pyautogui.press("down")
                last_action = current_time

            # PREVIOUS
            if fingers == [1,0,0,0,0] and current_time - last_action > 1:
                pyautogui.press("left")
                pyautogui.press("up")
                last_action = current_time

            # POINTER
            if fingers == [0,1,1,0,0]:
                x, y = lmList[8]
                cv2.circle(img, (x, y), 15, (0, 0, 255), cv2.FILLED)

            # DRAW ONLY LANDMARKS ON BLACK SCREEN
            mpDraw.draw_landmarks(img, handLms, mpHands.HAND_CONNECTIONS)

    cv2.imshow("PPT Controller", img)

    if cv2.waitKey(1) & 0xFF == 27:
        break
cap.release()
cv2.destroyAllWindows()
```
